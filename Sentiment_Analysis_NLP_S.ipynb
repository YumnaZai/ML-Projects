## Content description
- determines the sentiment expressed in a piece of text
- classified as positive or negative or neutral
- find a relationship between the features of a mobile phone (eg:- RAM, internal memory, etc.) and its selling price

### Load the DataSet

```python
import pandas as pd
import nltk
```

```python
df = pd.read_csv('Tweets.csv')
df.head(5)
```

```python
df = df[["airline_sentiment","text"]]
```

```python
df.shape
```

### Preprocess Text
- text by obtaining lower case
- removing URLs
- removing stop words
-  generating stem

```python
import nltk
import  string
import re
```

```python
from nltk.stem.porter import PorterStemmer

nltk.download('stopwords')
from nltk.corpus import stopwords

from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('punkt_tab')
ps = PorterStemmer()
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'http.?://[^\s]+[\s]?', '', text)
    text = nltk.word_tokenize(text)

    y = []
    for i in text:
      if i not in stopwords.words('english'):
          y.append(i)

    text = y[:]
    y.clear()

    for i in text:
      y.append(ps.stem(i))

    return" ".join(y)
```

```python
df["text_cleaned"] = df["text"].apply(clean_text)
```

```python
df.head()
```

```python
df.shape
```

### Feature Extraction

```python
# TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Create a tfid vector
tf_vec = TfidfVectorizer(max_features = 3000) # size of  the vocabularly

# generate TF-IDF vector representation of cleaned text in the column
# convert the vector representation to an array,
x = tf_vec.fit_transform(df['text_cleaned']).toarray()
 # transform from text to number array
```

```python
x.shape
```

```python
# to convert the column to an array
y = df['airline_sentiment'].values
```

```python
y.shape
```

### Train Model

### Split data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=2)
```

```python
#  Train a multinomial Nave Bayes classifier
from sklearn.naive_bayes import MultinomialNB # A ML model

model = MultinomialNB()
model.fit(x_train,y_train)
```

```python
# Measure accuracy
from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
# Train a multinomial Nave Bayes classifier
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
# Measure accuracy
y_pred = model.predict(x_test)
print(accuracy_score(y_test,y_pred))
```
